```r
library(dplyr)
library(sandwich)
library(lmtest)
library(boot)

# Read data
data <- read.csv("nhanes_bmi.csv")
```

```r
Y <- data$bmi
T <- data$School_meal

tau_dm <- mean(Y[T == 1]) - mean(Y[T == 0])

se_dm <- sqrt(
  var(Y[T == 1])/sum(T == 1) +
  var(Y[T == 0])/sum(T == 0)
)

tau_dm
se_dm
```

```r
model_or <- lm(
  bmi ~ treat + age + ChildSex + black + mexam +
    pir200_plus + WIC + FoodStamp + fsdchbi +
    AnyIns + RefSex + RefAge,
  data = data
)

summary(model_or)

# Robust SE
coeftest(model_or, vcov = vcovHC(model_or, type = "HC1"))
```

```r
coef(model_or)["treat"]
```

```r
model_or_int <- lm(
  bmi ~ treat * (
    age + ChildSex + black + mexam +
    pir200_plus + WIC + FoodStamp +
    fsdchbi + AnyIns + RefSex + RefAge
  ),
  data = data
)

summary(model_or_int)

coeftest(model_or_int, vcov = vcovHC(model_or_int, type = "HC1"))
```
