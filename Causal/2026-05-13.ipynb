```r
rand_comb <- function(n,x){
    choose(n, x)
    positions <- combn(n, x)
    
    result <- t(apply(positions, 2, function(pos) {
    vec <- rep(0, n)
    vec[pos] <- 1
    vec
    }))
    print(result)
}
```

```r
rand_comb(5, 3)
```
