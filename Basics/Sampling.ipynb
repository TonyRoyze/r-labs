# Rejection Sampling

```python
# target density
f <- function(x) (3/2) * (1 - x^2)

# proposal: Uniform(-1,1)
g <- function(x) 1/2
M <- 3  # scaling constant

N <- 10000
sample <- numeric(N)
attempts <- 0
for (i in 1:N) {
  repeat {
    attempts <- attempts + 1
    x <- runif(1, -1, 1)
    u <- runif(1)
    if (u <= f(x) / (M * g(x))) {
      sample[i] <- x
      break   # move to next sample
    }
  }
}
```
