# Mid Square Method

```r
mid_square <- function(seed, n) {
  for (i in 1:n) {
    rand <- seed * seed
    if (rand / 10000000 > 1) {
      mid <- as.integer(substr(as.character(rand), 3, 6))
    } else {
      mid <- as.integer(substr(as.character(rand), 2, 5))
    }
    print(mid / 10000)
    seed <- mid
  }
}

mid_square(1234, 20)
```

# L C G

```r
lcg <- function(seed, n, a, c, m) {
  for (i in 1:n) {
    rand <- (a * seed + c) %% m
    print(rand/m)
    seed <- rand
  }
}

mid_square(1234, 20)
```

# Period

```r
period <- function(x) {
  for (i in 1:(length(x) - 1)) {
    if (x[i] == x[length(x)]) {
      return(length(x) - i)
    }
  }
  return(NA)  # no repetition found
}
```
